# Extract Frames

```python
import cv2

def extract_frames_from_videos(video_dir, output_dir, skip_frames):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Iterate through each file in the video directory
    for filename in os.listdir(video_dir)[:]:
        print(filename)
        if filename.endswith(".mp4") or filename.endswith(".avi"):  # You can add more video extensions if needed
            video_path = os.path.join(video_dir, filename)
            video_name = os.path.splitext(filename)[0]

            # Open the video file
            vidcap = cv2.VideoCapture(video_path)
            frame_count = 0

            # Read the video frame by frame and save as images
            while True:
              ret, image = vidcap.read()
              if not ret:
                break
              if frame_count % skip_frames == 0:
                frame_path = os.path.join(output_dir, f"{video_name}_frame_{frame_count}.jpeg")
                image = cv2.resize(image, (0, 0), fx = 0.7, fy = 0.7)
                cv2.imwrite(frame_path, image)  # Save frame as JPEG
                success, image = vidcap.read()
              frame_count += 1
            os.remove(video_path)
            vidcap.release()
```

# Extract Faces from frames using YOLO from hugging face hub

```python
!pip install ultralytics
!pip install supervision
```

```python
# load libraries
from huggingface_hub import hf_hub_download
from ultralytics import YOLO
from supervision import Detections
from PIL import Image
import numpy as np

# download model
model_path = hf_hub_download(repo_id="arnabdhar/YOLOv8-Face-Detection", filename="model.pt")

# load model
model = YOLO(model_path)
```

```python
def crop_and_save(image_path):
  image = Image.open(image_path)
  output = model(image)
  results = Detections.from_ultralytics(output[0])
  bbox = str(results[0]).split("]")[0].replace(" ","").split("[")[-1].split(",")
  cx = float(bbox[0])
  cy = float(bbox[1])
  h = float(bbox[2])
  w = float(bbox[3])
  bbox = [cx, cy, h, w]
  image = cv2.imread(image_path)
  y2, y1, x2, x1 = bbox
  cropped_image = image[int(y1):int(y2), int(x1):int(x2)]
  file=image_path
  print(file)
  cv2.imwrite(file,cropped_image)
```
